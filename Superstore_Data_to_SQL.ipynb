```python
df.columns
df.columns = df.columns.str.lower() #This will convert column names in lower case
df.columns = df.columns.str.replace(' ','_')# This will remove space and add'_'
df.columns = df.columns.str.replace('-','_') # This will remove space and add'_'
df.columns
df.head()
```
